```python
import csv
import nltk
from nltk import FreqDist
import pandas as pd
import numpy as np
```

```python
## -- Step 1: tokenizing --
words = []
with open('Tweets2.csv', newline='', encoding="utf8") as csvfile:
	reader = csv.DictReader(csvfile, quoting=csv.QUOTE_ALL)
	for row in reader:
		#print(row['airline_sentiment'], row['text'])
		words = words + (nltk.tokenize.word_tokenize(row["text"].lower()))
print(words)
```

```python
len(words)
```

```python
## -- Step 2: stemming -- 
stemmedWords = []
from nltk.stem import PorterStemmer
pst=PorterStemmer()
for row in words:
	stemmedWords = stemmedWords + [pst.stem(row)]
	

print(stemmedWords)
```

```python
len(stemmedWords)
```

```python
## -- Step 3 A: stopword en-default --
from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
stoppedWord = []
for w in stemmedWords:
    if w not in stopWords:
        stoppedWord.append(w)
print(stoppedWord)
```

```python
list_stopword = []
## -- Step 3 B: Stopword additional
#masukkan semua stopwords ke file stopwords.txt
with open('stopwords.txt',encoding ='ISO-8859-1') as stopword:
	for line in stopword:
		list_stopword.append(line.strip())
cleanWords = []
for w in stoppedWord:
	if w not in list_stopword: 
		# jika kata tidak termasuk dlm stop words
		cleanWords.append(w) 
print(cleanWords)
```
